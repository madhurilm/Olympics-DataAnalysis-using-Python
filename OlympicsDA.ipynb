```python
#creating the dataset in which Medal column has notnull data
athlets_df_notnull=athlets_df[(athlets_df['Medal'].notnull())]
```

```python
athlets_df_notnull.Medal.unique()
```

```python
# Grouping by 'Team' and 'Medal', counting the occurrences
medal_counts = athlets_df_notnull.groupby(['Team', 'Medal']).size().unstack(fill_value=0)
# Adding a new column for total medals won by each country
medal_counts['Total'] = medal_counts.sum(axis=1)
# Sorting by total medals won
medal_counts_sorted = medal_counts.sort_values(by='Total', ascending=False).head(20)  # Top 20 countries
```

```python
medal_counts_sorted=medal_counts_sorted.drop('Total', axis=1)
medal_counts_sorted
```

#Plotting the Top 20 countries Medal distribution

```python
medal_counts_sorted.plot(kind='bar', color=['brown', 'Gold', 'grey'],figsize=(20,6))
plt.xlabel('Countries')
plt.title('Medal Distribution for top 20 countries')
```

```python
#Rio Olympics - 2016
team_names=athlets_df[(athlets_df.Year==2016) & (athlets_df.Medal=='Gold')].Team
team_names.value_counts().head(10)
```
